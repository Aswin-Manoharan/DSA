### Visualizations

```python
import matplotlib.pyplot as plt
```

```python
x = df.SL
y = df.SW
```

```python
plt.scatter(x,y)
plt.xlabel('SL')
plt.ylabel('SW')
plt.title('Scatter Plot of SL & SW')
plt.show()
```

```python
plt.hist(df['PL'],bins=30,color = 'y',edgecolor = 'r')
plt.xlabel('PL')
plt.ylabel('Frequency')
plt.title('Histogram of PL')
plt.show()
```

```python
plt.bar(df['Classification'],df['PW'],color='r',edgecolor='y')
plt.xlabel('Classification')
plt.ylabel('PW')
plt.title('Bar Plot of Classification and PW')
plt.xticks(rotation=45)
plt.show()
```
